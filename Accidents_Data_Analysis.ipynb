# Cleaning Data

## Cleaning missing values

* Percentage of missing values per column

```python
missing_values_percentage = (df.isna().sum().sort_values(ascending = False) / len(df))* 100
missing_values_percentage
```

```python
missing_values_percentage[missing_values_percentage != 0].plot(kind = 'barh')
plt.show()
```

* By observing above graph, there are some columns which are largely missing and they can be neglected because they may not affect our analysis

# Exploratory Data Analysis

Columns that we'll analysis
1. City
2. Start Time
3. Start Lat, Start Lng
4. Temperature
5. Weather Condition

```python
adf = df[['City', 'Start_Time', 'Start_Lat', 'Start_Lng', 'Temperature(F)', 'Weather_Condition']]
adf.head(5)
```

```python
# unique cities
cities = adf['City'].unique()
len(cities)
```

```python
# Number of accidents in different cities
cities_by_accident = df.City.value_counts()
cities_by_accident[:25]
```

```python
cities_by_accident[:25].plot(kind = 'barh')
```

* Among top 100 cities  in number of accidents, which states do they belong to most frequently

```python
high_acc_cities = cities_by_accident[cities_by_accident >= 1000]
low_acc_cities = cities_by_accident[cities_by_accident < 1000]
```

```python
# Percentage of High accident cities
(len(high_acc_cities)/len(cities))*100
```

```python
# Percentage of low accident cities
(len(low_acc_cities)/len(cities))*100
```

```python
#city with 1 accident
cities_by_accident[cities_by_accident == 1]
```

```python
sns.histplot(cities_by_accident, log_scale = True)
```

```python
sns.distplot(low_acc_cities)
```
